```python
batsman = pd.read_csv('player_stats.csv')
```

```python
plt.plot(batsman['Year'],batsman['V_Kohli'],
color='#D9F10F',linestyle='solid',linewidth=2,marker='>',markersize=6,label="Virat")
plt.plot(batsman['Year'],batsman['RG_Sharma'],
color='#FC00D6',linestyle='dashdot',linewidth=2,marker='<',markersize=6,label="Sharma")
plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
```
