```python
import csv

data = [
    [2008, 404, 165],
    [2009, 362, 246],
    [2010, 404, 307],
    [2011, 372, 557],
    [2012, 433, 364],
    [2013, 538, 639],
    [2014, 390, 359],
    [2015, 482, 505],
    [2016, 489, 973],
    [2017, 333, 308]
]


header = ['Year', 'RG_Sharma', 'V_Kohli']


with open('player_stats.csv', 'w', newline='') as file:
    writer = csv.writer(file)


    writer.writerow(header)


    writer.writerows(data)

print("CSV file created successfully!")
```

```python
plt.plot(batsman['Year'],batsman['V_Kohli'],color='#9B870C',linestyle = 'dashdot')
plt.plot(batsman['Year'],batsman['RG_Sharma'],color='#FC00D6',linestyle='dotted')

plt.title("Rohit Sharma Vs Virat Kohli ")

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['Year'],batsman['V_Kohli'],color='#9B870C',linestyle = 'dashdot',linewidth=2)
plt.plot(batsman['Year'],batsman['RG_Sharma'],color='#FC00D6',linestyle='dotted',linewidth=2)

plt.title("Rohit Sharma Vs Virat Kohli ")

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['Year'],batsman['V_Kohli'],color='#9B870C',linestyle = 'dashdot',linewidth=2,marker='<')
plt.plot(batsman['Year'],batsman['RG_Sharma'],color='#FC00D6',linestyle='dotted',linewidth=2,marker='>')

plt.title("Rohit Sharma Vs Virat Kohli ")

plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['Year'],batsman['V_Kohli'],color='#9B870C',linestyle = 'dashdot',linewidth=2,marker='<',markersize =7,label='Virat')
plt.plot(batsman['Year'],batsman['RG_Sharma'],color='#FC00D6',linestyle='dotted',linewidth=2,marker='>',markersize =6,label='Rohit')

plt.title("Rohit Sharma Vs Virat Kohli ")

plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```
