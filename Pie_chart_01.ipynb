```python
data = [
    {"batsman": "AB de Villiers", "batsman_runs": 31},
    {"batsman": "CH Gayle", "batsman_runs": 17},
    {"batsman": "R Rampaul", "batsman_runs": 12},
    {"batsman": "SS Tiwary", "batsman_runs": 2},
    {"batsman": "TM Dilshan", "batsman_runs": 33}
]

df = pd.DataFrame(data)

df.to_csv('gayle.csv', index=False)
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],
       autopct='%0.1f%%',
       colors=['orange','green','gold','orange','cyan'])

plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],
       autopct='%0.1f%%',
       colors=['orange','green','gold','orange','cyan'],
       explode = [0.02,0,0,0.07,0.09])

plt.show()
```
